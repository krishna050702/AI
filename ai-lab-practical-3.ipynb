# 8 puzzle solver

```python
# initial state
mainBoard = [['2','3','4'],['-','1','5'],['6','7','8']]
finalBoard = [['2','4','-'],['1','3','5'],['6','7','8']]
print(mainBoard)
```

```python
def displayBoard(board):
  for i in board:
    for j in i:
      print(j,end=" ")
    print("")
  print("")
displayBoard(mainBoard)
```

```python
def isGameOver(board,final):
  for i in range(0,len(board)):
    for j in range(0,len(board[0])):
      if(board[i][j] != final[i][j]):
        return False
  return True
isGameOver(mainBoard,finalBoard)
```

```python
# transistion functions
def findVoid(board):
  loc = [-1,-1]
  for i in range(0,len(board)):
    for j in range(0,len(board[0])):
      if(board[i][j] == '-'):
        loc[0] = i
        loc[1] = j
        return loc;
  return loc
findVoid(mainBoard)

def up(board):
  loc = findVoid(board)
  ls = [x[:] for x in board]
  if(loc[0] - 1 >= 0):
    ls[loc[0]][loc[1]] = ls[loc[0] - 1][loc[1]]
    ls[loc[0]-1][loc[1]] = '-'
  return ls
displayBoard(up(mainBoard))

def down(board):
  loc = findVoid(board)
  ls = [x[:] for x in board]
  if(loc[0] + 1 < len(ls)):
    ls[loc[0]][loc[1]] = ls[loc[0] + 1][loc[1]]
    ls[loc[0] + 1][loc[1]] = '-'
  return ls
displayBoard(down(mainBoard))

def left(board):
  loc = findVoid(board)
  ls = [x[:] for x in board]
  if(loc[1] - 1 >= 0):
    ls[loc[0]][loc[1]] = ls[loc[0]][loc[1] - 1]
    ls[loc[0]][loc[1] - 1] = '-'
  return ls
displayBoard(left(mainBoard))

def right(board):
  loc = findVoid(board)
  ls = [x[:] for x in board]
  if(loc[1] + 1 < len(ls)):
    ls[loc[0]][loc[1]] = ls[loc[0]][loc[1] + 1]
    ls[loc[0]][loc[1] + 1] = '-'
  return ls
displayBoard(right(mainBoard))
```

## Solving the game

```python
fn = [up,down,left,right]
mem = []

def inMem(board):
  for l in mem:
    if(isGameOver(l,board)):
      return True
  return False

ls = [mainBoard]

def letsGame():
  count = 0
  gen = 0
  while(len(ls) > 0):
    l = len(ls)

    # goal test
    for t in ls:
      if(isGameOver(t,finalBoard)):
        return "YES"
    count += 1
    print(str(count) + ": ",end="")
    print(ls)
    while(l > 0):
      if(inMem(ls[0]) == 1):
        ls.pop(0)
        l -= 1
        continue

      mem.append(ls[0])

      for f in fn:
        temp = f(ls[0])
        if(isGameOver(temp,finalBoard)):
          return "YES found on level: " + str(count) + " & steps: " + str(gen) + " !!"
        ls.append(f(ls[0]))
        gen += 1
    
      ls.pop(0)
      l -= 1
  return "NO"
letsGame()
```
